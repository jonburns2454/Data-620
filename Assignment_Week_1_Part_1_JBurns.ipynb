### Import packages

```python
import networkx as nx
import matplotlib.pyplot as plt 
import numpy as np 
```

### Make networkx graph
* choose fake nodes and edges just for the visualization

```python
G = nx.Graph()


#Node list
G.add_node("Jaylen Brown")
G.add_node("Jason Tatum")
G.add_node("Kristaps Porzingis")
G.add_node("Luka Doncic")
G.add_node("Kyrie Irving")
G.add_node("Kevin Durant")
G.add_node("Devin Booker")
G.add_node("Deandre Ayton")
G.add_node("Rob Williams")

#Edge list
G.add_edge("Jaylen Brown", "Jason Tatum")
G.add_edge("Kristaps Porzingis", "Luka Doncic")
G.add_edge("Kyrie Irving", "Kevin Durant")
G.add_edge("Devin Booker", "Deandre Ayton")
G.add_edge("Rob Williams", "Jaylen Brown")
G.add_edge("Rob Williams", "Jason Tatum")
G.add_edge("Rob Williams", "Deandre Ayton")
G.add_edge("Kyrie Irving", "Jaylen Brown")
G.add_edge("Kyrie Irving", "Jaylen Brown")
G.add_edge("Kyrie Irving", "Rob Williams")
G.add_edge("Kyrie Irving", "Luka Doncic")
G.add_edge("Kristaps Porzingis", "Jaylen Brown")
G.add_edge("Kristaps Porzingis", "Jason Tatum")

nx.draw(G, with_labels=True)
plt.show()
```

```python
nx.draw_circular(G, with_labels=True)

# Show the plot
plt.show()
```

Just off the top of my head I wanted to look at my favorite Celtics players from the last few years and graph their overlap with other big names in the league. In addition I got to play around with the network x functions. Based simply on readability I think that the draw_circular graph looks much better than the autogerated one from .draw().
